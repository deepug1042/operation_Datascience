```python
import pandas as pd
import numpy as np
```

```python
data = r'C:\Users\Dell\Desktop\employee.json' #Reading the json file which i have created in notepad using key and values(multiple values in list)
# r--->It is used in the above case because the file path should be treated as raw string if 'r' is not used then we have to replace single slash with double back slash
# In orderwords using 'r' makes the file path cleaner and easier to write.
#with 'r'--> single back slash, without 'r'-->double back slash
```

```python
#convert json file to different json format
print(df.to_json())
#if you see above for every key value there is '0' along with it so ,to avoid this we use orient parameter setting to "records"
print(df.to_json(orient="records"))
#now the data is according to the values to the records 
```

```python
df.to_json(orient='split')
#it actually splits the key and corresponding value sepeartely
```

```python
df1.to_json(orient="records")
#it is important to the key values in records
```

# Reading HTML contents
-->It means that we are basically reading table from an html page
-->It is also a type of web scrapping bt with respective table within a webpage

```python
url = 'https://www.fdic.gov/bank/individual/failed/banklist.html'
hd = pd.read_html(url)
#the read_html picks the thead tags and grabs all the list of tables from the webpage  
```

```python
hd
```

```python
type(hd)
```

```python
hd[0]
#it doesnt care about the connceted links it will only display the records which are presented within the webpage
```
