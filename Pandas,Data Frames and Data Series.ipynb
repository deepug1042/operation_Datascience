# what is a Data Frame
    A dataframe is a combination of rows and columns which represents the data in a format that shows how the data look like

```python
#creating an dataframe manually including the rows and columns
df = pd.DataFrame(np.arange(0,20).reshape(5,4),index = ['Row1','Row2','Row3','Row4','Row5'],columns = ['column1','column2','column3','column4'])
#here we take a numbers from 0 to 20 and rshape to 5 rows and 4 columns
#note: that the index should be of same rows of reshaped value 
```

```python
df
```

```python
#to convert the dataframe to csv file we use to_csv inbuilt function
df.to_csv('1st_csv')
```

# Data series
    Data series is the values of a single row ie only one record is known as series if more tha  one then it is dataframe

# accesing elements is of two type:
    1) loc[]  ----> simple location
    2) iloc[]  -->index location

```python
df
```

```python
#loc[]----
df.loc['Row1']  #this is known as series
print(type(df.loc['Row1']))
```

```python
# iloc[]--->
print(df.iloc[0:3,0:2])
print(type(df.iloc[0:3,0:2]))
#here if we check the type of the function we will get as dataframe
#so a single row values is known as data series and mutiple values with multiple columns are called as dataframe
```

```python
#converting dataframes to arrays
df.iloc[:,1:].values 
#we generally use values() function to convert the datframe to arrays
```

```python
from io import StringIO , BytesIO
```

```python
#Here we are creating an dataset of 3 columns which we write in the form of a string later we are reading the data by using stringio function 
data = ('column1,column2,column3\n'
               'a,b,c\n'
                'd,e,e\n'
               'a,n,a')
```

```python
type(data)
#The datatype of the data created is string
```

```python
pd.read_csv(StringIO(data)) #the data which was created is displayed in an dataframe format ie in rows and columns
#StringIO generally craetes a text memory buffer for the values given
#read_csv-->actuallly reads the data as comma sepearted values
```

```python
#Read from a specific column
data = pd.read_csv(StringIO(data),usecols = ['column1','column2'])
```

```python
data = ('index,a,b,c\n'
       '4,apple,orange,5.7\n'
       '8,strawberry,mango,10')
```

```python
pd.read_csv(StringIO(data))
#here already default index such as 0 and 1 is assigned bt we need to add the given values as indexes to the row we do it by
```

```python
pd.read_csv(StringIO(data),index_col = 0) #by putting index_col=0 we put the defined values to the row index
```

```python
pd.read_csv(StringIO(data),index_col = 1) #the values will change accordingly
```

```python
data = ('a,b,c\n'
       '4,apple,orange,\n'
       '8,strawberry,mango,')
```

```python
pd.read_csv(StringIO(data))
#4,8 are considered as index as thy are numbers and the default datatype is none
```

```python
#To solve the above probelm we use index_col=False
pd.read_csv(StringIO(data),index_col=False)
```

```python
#combining usecols and index_cols together
pd.read_csv(StringIO(data),usecols = ['b','c'],index_col=False)
```
